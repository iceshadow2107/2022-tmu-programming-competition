```python
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline
plt.rcParams['font.sans-serif']=['Microsoft YaHei'] # 用來正常顯示中文標籤
plt.rcParams['axes.unicode_minus']=False # 用來正常顯示負號
from datetime import datetime
plt.figure(figsize=(16,10))
import pyecharts.options as opts
from pyecharts.charts import Line
from pyecharts.faker import Faker
from pyecharts.charts import Bar
import os
from pyecharts.options.global_options import ThemeType
```

## 陳時中

```python
chen_china = pd.read_csv("./政治人物csv/陳時中/陳時中(中時_label).csv", encoding = 'utf-8')
chen_china = chen_china.assign(month = "")
chen_china.head()
```

```python
chen_china['month'] = chen_china.TIME.map(lambda x: x[2:7])
```

```python
chen_china = chen_china.sort_values(by=['month'])
chen_china.head(30)
```

## 按照label各個值進行分類

```python
chen_china_n = chen_china[chen_china['label'] == '中性']
chen_china_n.head()
```

```python
chen_china_n.to_csv('./政治人物csv/陳時中/陳時中_中時_中性.csv', index = False)
```

```python
chen_china_f = chen_china[chen_china['label'] == '正面']
chen_china_f.head()
```

```python
chen_china_f.to_csv('./政治人物csv/陳時中/陳時中_中時_正面.csv', index = False)
```

```python
chen_china_b = chen_china[chen_china['label'] == '負面']
chen_china_b.head()
```

```python
chen_china_b.to_csv('./政治人物csv/陳時中/陳時中_中時_負面.csv', index = False)
```

## 用groupby做分組 

```python
freq = chen_china.groupby(['month', 'label']).count() 
freq
```

```python
freq_n = chen_china_n.groupby(['month', 'label']).count() 
freq_n
```

```python
freq_f = chen_china_f.groupby(['month', 'label']).count() 
freq_f
```

```python
freq_b = chen_china_b.groupby(['month', 'label']).count() 
freq_b
```

## 單純標顏色

```python
(freq.style.applymap(lambda x: 'background-color: rgb(153, 255, 51)', subset = pd.IndexSlice[:, 'TIME']))
```

## 進行視覺化

```python
x = ['20-10', '20-11', '20-12', '21-01', '21-02', '21-03', '21-04', '21-05', '21-06', '21-07', '21-08', 
     '21-09', '21-10', '21-11', '21-12', '22-01', '22-02', '22-03', '22-04']

line = (
    Line()
        .add_xaxis(xaxis_data = x) # X軸
        .add_yaxis('正面', freq_f.cleaned_text.tolist())
        .add_yaxis('中性', freq_n.cleaned_text.tolist())
        .add_yaxis('負面', freq_b.cleaned_text.tolist())
        .set_global_opts(title_opts=opts.TitleOpts(title="陳時中_中時")) #標題
    )

line.render()#_notebook()
# init_opts=opts.InitOpts(theme=ThemeType.DARK)
```
