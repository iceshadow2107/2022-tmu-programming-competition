### 去除贅字與標點符號

```python
def clean_seg_list(seg_list):
    stop_words = [line.strip() for line in open
                  (stop_words_file, encoding="utf-8").readlines()]
    return [seg for seg in seg_list if seg[0] not in stop_words]
```

```python
def get_cleaned_seg(df):    
    cleaned = []
    for line in df['cut_text']:
        items = line.split()
        item = ' '.join(clean_seg_list(items))
        cleaned.append(item)
    df['cleaned_text'] = cleaned
    
    return df
```

```python
from monpa import utils

AZ_china_new.fillna('',inplace=True)

for i in range(len(AZ_china_new)):
    result = []
    for item in utils.short_sentence(AZ_china_new.loc[i,'TITLE'] + AZ_china_new.loc[i,'CONTENT']):
        if item:
            result.append(monpa.cut(item))

    tmp_result =""
    for k in range(len(result)):
        tmp_result += " ".join(result[k])
    AZ_china_new.loc[i,'cut_text'] = tmp_result
    print('i=:',i)
    #break

AZ_china_new.head(), len(AZ_china_new)
```
