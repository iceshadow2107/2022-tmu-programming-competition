### 去除贅字與標點符號

```python
def clean_seg_list(seg_list):
    stop_words = [line.strip() for line in open
                  (stop_words_file, encoding="utf-8").readlines()]
    return [seg for seg in seg_list if seg[0] not in stop_words]
```

```python
def get_cleaned_seg(df):    
    cleaned = []
    for line in df['cut_text']:
        items = line.split()
        item = ' '.join(clean_seg_list(items))
        cleaned.append(item)
    df['cleaned_text'] = cleaned
    
    return df
```
